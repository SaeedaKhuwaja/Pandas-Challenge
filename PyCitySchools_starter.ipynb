```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
unique_schools = school_data_complete["school_name"].nunique()
print(unique_schools)
```

```python
# Calculate the total number of students
total_students = school_data_complete["student_name"].count()
print(total_students)
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
print(total_budget)
```

```python
# Calculate the average (mean) math score
math_average = school_data_complete["math_score"].mean()
print(math_average)
```

```python
# Calculate the average (mean) reading score
reading_average = school_data_complete["reading_score"].mean()
print(reading_average)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
math_percentage = (school_data_complete[school_data_complete['math_score'] >= 70].math_score).count() / total_students * 100
print(math_percentage)
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
reading_percentage = (school_data_complete[school_data_complete['reading_score'] >= 70].reading_score).count() / total_students * 100
print(reading_percentage)
```

```python
# Use the following to calculate the percentage of students that passed math and reading
math_reading_percentage = len(school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]) / total_students * 100
print(math_reading_percentage)
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
District_Summary = pd.DataFrame({
    'Total Schools' : [unique_schools],
    'Total Students': [total_students],
    'Total Budget' :  [total_budget],
    'Average Math Score' : [math_average],
    'Average Reading Score' : [reading_average],
    '% Passing Math' : [math_percentage],
    '% Passing Reading' : [reading_percentage],
    'Overall Percentage' : [math_reading_percentage]
})

# Formatting
District_Summary['Total Budget'] = District_Summary['Total Budget'].map("${:,.0f}".format)
District_Summary['Total Students'] = District_Summary['Total Students'].map("{:,.0f}".format)

# Display the DataFrame
District_Summary
```

## School Summary

```python
# Use the code provided to select the school type
school_summary = school_data.set_index(['school_name'])["type"]
```

```python
# Calculate the total student count
students_count = school_data_complete["school_name"].value_counts()
print(students_count)
```

```python
# Calculate the total school budget and per capita spending
school_budget = school_data.groupby(by="school_name")["budget"].mean()
print(school_budget)

# per capita spending
per_student_budget =  school_budget / students_count
print(per_student_budget)
```

```python
# Calculate the average test scores
average_score = student_data.groupby(by="school_name").mean()

#setting the variables
math_score = average_score["math_score"]
reading_score = average_score["reading_score"]
print(math_score)
print(reading_score)
```

```python
# Calculate the number of schools with math scores of 70 or higher
passing_math = school_data_complete.groupby("school_name")['math_score'].apply(lambda x: (x >= 70).sum()) / students_count * 100
print(passing_math)
```

```python
# Calculate the number of schools with reading scores of 70 or higher
passing_reading = school_data_complete.groupby("school_name")['reading_score'].apply(lambda x: (x >= 70).sum()) / students_count * 100
print(passing_reading)
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_count = school_data_complete[(school_data_complete['math_score'] >= 70) & 
                                        (school_data_complete['reading_score'] >= 70)].groupby("school_name").count()
passing_count["reading_score"]
```

```python
# Use the provided code to calculate the passing rates
overall_passing = passing_count['reading_score'] / students_count * 100
overall_passing
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
Per_School_Summary = pd.DataFrame(
{
    "School Type" : school_summary,
    'Total Students': students_count,
    'Total School Budget' : school_budget,
    'Per Student Budget' : per_student_budget,
    'Average Math Score' : math_score,
    'Average Reading Score' : reading_score,
    '% Passing Math' : passing_math,
    '% Passing Reading' : passing_reading,
    '% Overall Passing' : overall_passing
})
# Formatting
Per_School_Summary['Total School Budget'] = Per_School_Summary['Total School Budget'].map("${:,.2f}".format)
Per_School_Summary['Per Student Budget'] = Per_School_Summary['Per Student Budget'].map("${:,.0f}".format)

# Display the DataFrame
Per_School_Summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
highest_performing_schools = Per_School_Summary.sort_values(["% Overall Passing"], ascending = False)
highest_performing_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_performing_schools = Per_School_Summary.sort_values(["% Overall Passing"], ascending = True)
bottom_performing_schools.head(5)
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
#student_data 
nineth_math = student_data.loc[student_data["grade"] == "9th"]
tenth_math = student_data.loc[student_data["grade"] == "10th"]
eleventh_math = student_data.loc[student_data["grade"] == "11th"]
twelveth_math = student_data.loc[student_data["grade"] == "12th"]

# Group by "school_name" and take the mean of each.
nineth_math = nineth_math.groupby("school_name")["math_score"].mean()
tenth_math = tenth_math.groupby("school_name")["math_score"].mean()
eleventh_math = eleventh_math.groupby("school_name")["math_score"].mean()
twelveth_math = twelveth_math.groupby("school_name")["math_score"].mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
    "9th" : nineth_math,
    "10th" : tenth_math,
    "11th" : eleventh_math,
    "12th" : twelveth_math
})

# Minor data wrangling
math_scores_by_grade['9th'] = math_scores_by_grade['9th'].map("{:,.1f}".format)
math_scores_by_grade['10th'] = math_scores_by_grade['10th'].map("{:,.1f}".format)
math_scores_by_grade['11th'] = math_scores_by_grade['11th'].map("{:,.1f}".format)
math_scores_by_grade['12th'] = math_scores_by_grade['12th'].map("{:,.1f}".format)

# Display the DataFrame
math_scores_by_grade
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
nineth_reading = student_data.loc[student_data["grade"] == "9th"]
tenth_reading = student_data.loc[student_data["grade"] == "10th"]
eleventh_reading = student_data.loc[student_data["grade"] == "11th"]
twelveth_reading = student_data.loc[student_data["grade"] == "12th"]

# Group by "school_name" and take the mean of each.
nineth_reading = nineth_reading.groupby("school_name")["reading_score"].mean()
tenth_reading = tenth_reading.groupby("school_name")["reading_score"].mean()
eleventh_reading = eleventh_reading.groupby("school_name")["reading_score"].mean()
twelveth_reading = twelveth_reading.groupby("school_name")["reading_score"].mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({
    "9th" : nineth_reading,
    "10th" : tenth_reading,
    "11th" : eleventh_reading,
    "12th" : twelveth_reading
})

# Minor data wrangling
reading_scores_by_grade['9th'] = reading_scores_by_grade['9th'].map("{:,.1f}".format)
reading_scores_by_grade['10th'] = reading_scores_by_grade['10th'].map("{:,.1f}".format)
reading_scores_by_grade['11th'] = reading_scores_by_grade['11th'].map("{:,.1f}".format)
reading_scores_by_grade['12th'] = reading_scores_by_grade['12th'].map("{:,.1f}".format)

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 584.9, 629.9, 644.9, 679.9]
group_name = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
Scores_by_spending = Per_School_Summary.copy()
Scores_by_spending["Per Student Budget"] = Scores_by_spending["Per Student Budget"].str.replace('$','').astype(int)
print(Scores_by_spending["Per Student Budget"])
```

```python
# Use `pd.cut` to categorize spending based on the bins.
Scores_by_spending["Spending Ranges"] = pd.cut(Scores_by_spending["Per Student Budget"], spending_bins, labels = group_name)
Scores_by_spending
```

```python
#  Calculate averages for the desired columns. 
math_averages = Scores_by_spending.groupby("Spending Ranges")["Average Math Score"].mean()
reading_averages = Scores_by_spending.groupby("Spending Ranges")["Average Reading Score"].mean()
math_percent = Scores_by_spending.groupby("Spending Ranges")["% Passing Math"].mean()
reading_percent = Scores_by_spending.groupby("Spending Ranges")["% Passing Reading"].mean()
overall_percent = Scores_by_spending.groupby("Spending Ranges")["% Overall Passing"].mean()

print(math_averages)
print(reading_averages)
print(math_percent)
print(reading_percent)
print(overall_percent)
```

```python
# Assemble into DataFrame
spending_ranges = pd.DataFrame({
    "Average Math Score" : math_averages,
    "Average Reading Score" : reading_averages,
    "% Passing Math" : math_percent,
    "% Passing Reading" : reading_percent,
    "% Passing Overall" : overall_percent
})

# Display results
spending_ranges
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 999.9, 1999.9, 4999.9]
group_name = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
Scores_by_size = Per_School_Summary.copy()

# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
Scores_by_size["School Size"] = pd.cut(Scores_by_size["Total Students"], size_bins, labels = group_name)
Scores_by_size
```

```python
# Calculate averages for the desired columns. 
math_size = Scores_by_size.groupby("School Size")["Average Math Score"].mean()
reading_size = Scores_by_size.groupby("School Size")["Average Reading Score"].mean()
math_percent_size = Scores_by_size.groupby("School Size")["% Passing Math"].mean()
reading_percent_size = Scores_by_size.groupby("School Size")["% Passing Reading"].mean()
overall_percent_size = Scores_by_size.groupby("School Size")["% Overall Passing"].mean()

print(math_size)
print(reading_size)
print(math_percent_size)
print(reading_percent_size)
print(overall_percent_size)
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({
    "Average Math Score" : math_size,
    "Average Reading Score" : reading_size,
    "% Passing Math" : math_percent_size,
    "% Passing Reading" : reading_percent_size,
    "% Passing Overall" : overall_percent_size
})

# Display results
size_summary
```
